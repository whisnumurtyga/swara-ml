```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, array_to_img, load_img
import numpy as np
```

```python
def augment_images(directory, num_augmentations):
    # Define the augmentation parameters
    datagen = ImageDataGenerator(
        rotation_range=3,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    # Iterate through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('png', 'jpg', 'jpeg')):
                img_path = os.path.join(root, file)
                img = load_img(img_path)
                x = img_to_array(img)
                x = np.expand_dims(x, axis=0)
                
                # Generate augmentations
                i = 0
                for batch in datagen.flow(x, batch_size=1):
                    aug_img = array_to_img(batch[0])
                    aug_img.save(os.path.join(root, f'augmentasi{i}_{file}'))
                    i += 1
                    if i >= num_augmentations:
                        break

# Example usage
augment_images('./dataset', 5)
```

```python
def remove_augmented_images(directory):
    # Iterate through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Check if the filename contains 'augmentasi'
            if 'augmentasi' in file:
                file_path = os.path.join(root, file)
                os.remove(file_path)
                print(f"Removed file: {file_path}")

# Example usage
remove_augmented_images('dataset')
```
